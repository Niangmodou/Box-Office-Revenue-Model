```
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from google.colab import drive
from sklearn.linear_model import LinearRegression
drive.mount('/content/drive')
```

```
PATH = "/content/drive/My Drive/Colab Notebooks/MovieBudget LinearRegression/cost-revenue-clean.csv"
dataFrame = pd.read_csv(PATH)
```

```
dataFrame.describe()
```

```
feature = DataFrame(dataFrame, columns = ['production_budget_usd'])
target = DataFrame(dataFrame, columns = ['worldwide_gross_usd'])
```

```
plt.figure(figsize=(10,6))
plt.scatter(feature,target,alpha=0.5)
plt.title("Production Budget and Revenue")
plt.xlabel("Production Budget($)")
plt.ylabel("Worldwide Revenue(S)")
plt.xlim(0,4.5e8)
plt.ylim(0,3e9)
plt.show()
```

```
regressor = LinearRegression()
regressor.fit(feature,target)
```

### Slope Coefficient and Intercept

> Indented block



```
regressor.coef_
```

```
regressor.intercept_
```

### Plotting line of best fit

```
plt.figure(figsize=(10,6))
plt.scatter(feature,target,alpha=0.5)
plt.plot(feature,regressor.predict(feature), color="red",linewidth=3)

plt.title("Production Budget and Revenue")
plt.xlabel("Production Budget($)")
plt.ylabel("Worldwide Revenue(S)")
plt.xlim(0,4.5e8)
plt.ylim(0,3e9)
plt.show()
```

## Evaluating the Model's performance

```
regressor.score(feature,target) * 100
```
